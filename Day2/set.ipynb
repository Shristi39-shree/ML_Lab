1 : write a program to demonstrate data preprocessing steps - like handling missing values, encoding categorical data and features scaling

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer     # For handling missing values
from sklearn.preprocessing import LabelEncoder, StandardScaler  # For encoding + scaling
#LabelEncoder -> converts text labels (like "Iris-setosa") into numbers (0,1,2).
# StandardScaler -> scales numeric features to a standard range (mean = 0, std = 1).

# 1. Load the Iris dataset
df = pd.read_csv(
    "iris.data.csv",
    header=None,   # tells pandas there is NO header line in the file
    names=["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]  # set column names
)

print("Original data (first 5 rows):")
print(df.head())  # Show first 5 rows

# 2. Separate numeric and categorical columns
num_cols = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]  # numeric features
cat_cols = ["Species"]   # categorical feature

# 3. Handle missing values in numeric columns using mean
num_imputer = SimpleImputer(strategy="mean") # create imputer for numbers
df[num_cols] = num_imputer.fit_transform(df[num_cols])  # apply to numeric columns

# 4. Handle missing values in categorical column  using most frequent
cat_imputer = SimpleImputer(strategy="most_frequent")   # imputer for category
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])  # apply to Species

print("\nAfter handling missing values:")
print(df.head())

# 5. Encode categorical column 'Species' into numbers
label = LabelEncoder()                      # create label encoder
df["Species"] = label.fit_transform(df["Species"])  # Iris-setosa/versicolor/virginica -> 0/1/2

print("\nAfter encoding 'Species':")
print(df.head())

# 6. Feature scaling (standardization) for numeric columns
scaler = StandardScaler()                   # create scaler
df[num_cols] = scaler.fit_transform(df[num_cols])  # scale numeric columns

print("\nAfter feature scaling:")
print(df.head())   # final preprocessed data
```

2 : Plot the distribution of a feature using matplotlib library

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (3, 2))
sns.histplot(df['SepalLength'], bins=20)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (standardized)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize = (3, 2))
sns.histplot(df['SepalWidth'], bins=20, color='orange')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (standardized)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(3,2))
sns.histplot(df['PetalLength'], bins=20, color='green')
```

3 : Create scatter plots to identify relationship between features using the scatter plot function.

```python
plt.figure(figsize=(6,4))
plt.scatter(df["SepalLength"], df["SepalWidth"], c="blue")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()

plt.figure(figsize = (4,3))
plt.scatter(df['PetalLength'], df['PetalWidth'], c = 'cyan')
plt.ylabel("Petal Width")
plt.xlabel("Petal Length")
plt.show()
```

4 : Use a correlation heatmap to find the relation between multiple features with the heatmap function.

```python
# Calculate correlation matrix for all numeric columns
corr_matrix = df.corr()

plt.figure(figsize=(8, 6))

# Draw a heatmap of the correlation matrix
# annot=True → show correlation numbers inside cells
# cmap="coolwarm" → color range from blue (low) to red (high)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap of Iris Dataset Features")

plt.show()
```
